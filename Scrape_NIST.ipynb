```python
import requests # requests html
from bs4 import BeautifulSoup # parse HTML
import cirpy # api made to get all the info you need (requests from internet and parse html)
import urllib # requests html
# display items
from IPython.display import HTML # display HTML in jupyter notebook
```

```python
%%time
acetylene = cirpy.resolve('acetylene','cas')
```

```python
print(acetylene)
```

```python
cas = acetylene[-1] # pick last one, it seems to be the correct one we want to look at
```

```python
%%time
url = 'https://webbook.nist.gov/cgi/cbook.cgi?ID=C%g&Units=SI'%int(cas.replace('-',''))
#url = 'https://en.wikipedia.org/wiki/Shaun_Harris'
page = requests.get(url,headers={'Connection': 'close'})
```

```python
display(HTML(page.text))
```

```python
soup=BeautifulSoup(page.text,'html.parser')
for i,li in enumerate(soup.select('li')):
    if 'Molecular weight' in li.text:
        print(i,li.text)
```
